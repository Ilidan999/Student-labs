```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

from IPython.core.pylabtools import figsize
```

Задание №1


```python
# Загрузка данных
diamonds = pd.read_csv("data/diamonds.csv")

plt.figure(figsize=(8, 4))
sns.scatterplot(data=diamonds, x='price', y='depth')
plt.title('Зависимость глубины бриллианта от его цены')

# Построение violin plot по цене для разных типов огранки
order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']

plt.figure(figsize=(10, 6))
sns.violinplot(data=diamonds, x='cut', y='price',order=order)
plt.title('Распределение цены бриллиантов по типу огранки')
plt.show()
```

Задание №2

```python
mtcars = pd.read_csv("data/mtcars.csv")

# Построение графика плотности
plt.figure(figsize=(10, 6))
sns.kdeplot(data=mtcars, x='hp', hue='carb', warn_singular=False,  palette="Set2")
plt.title("График плотности")



#Ящик с усами
plt.figure(figsize=(6, 6))
sns.boxplot(data=mtcars, x=mtcars["cyl"].astype(str), y="disp")
plt.title("Зависимость disp от cyl")


plt.show()
```

Задание №3-А

```python
# Загрузка данных
df = pd.read_csv('data/ChickWeight.csv')

#Данные
data = Counter(df['Diet'])
labels = data.keys()
sizes = data.values()

# Построение круговой диаграммы
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Распределение цыплят по типам диеты')


# Построение гистограммы
plt.figure(figsize=(6, 6))
plt.hist(df['weight'])
plt.title('Гистограмма распределения веса цыплят')


#Точечная диаграмма
plt.figure(figsize=(6, 6))
sns.scatterplot(data=df, x="Diet", y="Chick")
plt.title('Точечная диаграмма по признакам Diet и Chick')


#Ящик с усами
plt.figure(figsize=(6, 6))
sns.boxplot(data=df, x="Diet", y="Chick")
plt.title("Зависимость Chick от Diet")
plt.show()


#Данные
data = df.corr()

#Тепловая карта
plt.figure(figsize=(7,6))
sns.heatmap(data,annot=True, fmt=".2f")
plt.title('Корреляции признаков')


plt.show()
```

Задание №3-Б

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Diet', y='weight')
plt.title('Распределение Diet по weight')


plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Diet", y="id", hue="weight", marker='o')
plt.title("Изменение Diet по id (по weight)")


plt.show()
```
